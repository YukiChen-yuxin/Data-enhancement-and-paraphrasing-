### 把未登录词字符删掉

```python
import json
```

```python
with open('C:/Users/yuki/Desktop/大创/数据/结果/bilstm+att结果汇总.json',encoding="utf-8") as file_obj:
    result_lst=json.load(file_obj)
```

```python
def isChinese(word):
    for ch in word:
        if not '\u4e00' <= ch <= '\u9fff':
            return False
    return True
```

```python
for i in result_lst:
    i['复述句子']=i['复述句子'].replace('<UNK>','')
    i['复述句子']=i['复述句子'].replace('?','？')
    i['复述句子']=i['复述句子'].replace('!','！')
    i['复述句子']=i['复述句子'].replace('.','，')
    temp=i['复述句子']
    if temp != '':
        if isChinese(temp[-1])==False:
            i['复述句子']=temp[:-1]
```

```python
f=open('C:/Users/yuki/Desktop/大创/数据/新数据/测试参考句子.txt','r',encoding='utf-8')
#测试参考句子
test_lst2=[]
lines = f.readlines()      #读取全部内容 ，并以列表方式返回
for line in lines:
    line=line.replace('\n','')
    line=line.replace('\t','')
    test_lst2.append(line)
```

```python
f=open('C:/Users/yuki/Desktop/大创/数据/新数据/测试原句子.txt','r',encoding='utf-8')
#测试参考句子
test_lst1=[]
lines = f.readlines()      #读取全部内容 ，并以列表方式返回
for line in lines:
    line=line.replace('\n','')
    line=line.replace('\t','')
    test_lst1.append(line)
```

```python
import jieba
```

```python
#测试参考句子分词
test_lst2_jieba=[]
for w in test_lst2:
    word = " ".join(word for word in jieba.lcut(w))
    test_lst2_jieba.append(word)
```

```python
count=0
result_all=[]
for temp_dic in result_lst:
    tt={}
    if temp_dic['原句子']==test_lst1[count]:
        tt['原句子']=temp_dic['原句子']
        tt['原句子分词']=temp_dic['原句子分词']
        tt['参考句子']=test_lst2[count]
        tt['参考句子分词']=test_lst2_jieba[count]
        tt['复述句子']=temp_dic['复述句子']
        fushu=" ".join(word for word in jieba.lcut(temp_dic['复述句子']))
        tt['复述句子分词']=fushu
        
    count+=1
    result_all.append(tt)
```

```python
filename='C:/Users/yuki/Desktop/大创/数据/结果/bilstm+att结果汇总all.json'
with open(filename,'w',encoding='utf-8') as file_obj:
    json.dump(result_all,file_obj,ensure_ascii=False,indent = 4)
```
